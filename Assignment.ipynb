```python
#sampling 
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("sampling.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
bins = np.linspace(min(df['Monthly Family Income']), max(df['Monthly Family Income']),5,endpoint=True)
groupNames = ["Stratum_1", "Stratum_2", "Stratum_3","Stratum_4"]
df['Strata'] = pd.cut(df['Monthly Family Income'], bins, labels = groupNames, include_lowest = True)
df.head()
```

```python
bins
```

```python
Stratum_1=df[df['Strata']=='Stratum_1']
Stratum_1
```

```python
Stratum_2=df[df['Strata']=='Stratum_2']
Stratum_2
```

```python
Stratum_3=df[df['Strata']=='Stratum_3']
Stratum_3
```

```python
Stratum_4=df[df['Strata']=='Stratum_4']
Stratum_4
```

```python
#exporting_file
Stratum_1.to_excel(r'Stratum_1.xlsx', index=False)
Stratum_2.to_excel(r'Stratum_2.xlsx', index=False)
Stratum_3.to_excel(r'Stratum_3.xlsx', index=False)
Stratum_4.to_excel(r'Stratum_4.xlsx', index=False)
```

```python
#measuring the porpotion
population_size=49
sample_size=10
sample_propotion=sample_size/population_size
sample_propotion
```

```python
#measuring the population in stratum
Stratum_pouplation_2=19
Stratum_pouplation_1=15
Stratum_pouplation_4=10
Stratum_pouplation_3=5
```

```python
stratum_1=round(Stratum_pouplation_1*sample_propotion)
stratum_2=round(Stratum_pouplation_2*sample_propotion)
stratum_3=round(Stratum_pouplation_3*sample_propotion)
stratum_4=round(Stratum_pouplation_4*sample_propotion)
```

```python
print(f"**Propotional Allocation**\nstratum_1= {stratum_1}\nstratum_2= {stratum_2}\nstratum_3= {stratum_3}\nstratum_4= {stratum_4}")
```

```python
# random sampling in Stratum_1
stratum_1=Stratum_1.sample(n = 3,random_state = 17)
stratum_1
```

```python
# random sampling in Stratum_2
stratum_2=Stratum_2.sample(n =4,random_state = 17)
stratum_2
```

```python
# random sampling in Stratum_3
stratum_3=Stratum_3.sample(n =1,random_state = 17)
stratum_3
```

```python
# random sampling in Stratum_4
stratum_4=Stratum_4.sample(n =2,random_state = 2)
stratum_4
```

```python
#exporting_file
stratum_1.to_excel(r'sample_stratum_1.xlsx', index=False)
stratum_2.to_excel(r'sample_stratum_2.xlsx', index=False)
stratum_3.to_excel(r'sample_stratum_3.xlsx', index=False)
stratum_4.to_excel(r'sample_stratum_4.xlsx', index=False)
```
